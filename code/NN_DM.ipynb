# Plotting Neural Network with Dimensinality Reduction (PCA, ICA, RP, RF)


```
from matplotlib import pyplot as plt
import matplotlib.lines as mlines
```

```
hidden_layer_size = [2, 4, 6, 8, 10]
#PCA
pca_train_score = [0.5211842776927004, 0.5395610005104645, 0.5438999489535478, 0.552833078101072, 0.5584481878509444]
pca_test_score = [0.4806122448979592, 0.49081632653061225, 0.4989795918367347, 0.5020408163265306, 0.5071428571428571]

#ICA
ica_train_score = [0.47320061255742724, 0.5040837161817253, 0.46809596733027053, 0.5301174068402246, 0.5260336906584993]
ica_test_score = [0.4520408163265306, 0.47551020408163264, 0.4520408163265306, 0.5071428571428571, 0.5061224489795918]

# RP
rp_train_score = [0.5454313425216948, 0.5451761102603369, 0.5477284328739153, 0.5546197039305768, 0.557427258805513]
rp_test_score = [0.5040816326530613, 0.5112244897959184, 0.5122448979591837, 0.5163265306122449, 0.5448979591836735]

# RF
rf_train_score = [0.5413476263399694, 0.5334354262378764, 0.5474732006125574, 0.5436447166921899, 0.544920877998979]
rf_test_score = [0.5040816326530613, 0.5081632653061224, 0.5071428571428571, 0.5173469387755102, 0.5163265306122449]

# Plot graph
plt.title('NN with Dim Red - Wine Dataset')
plt.xlabel('Hidden layer size')
plt.ylabel('Accuracy score')

plt.plot(hidden_layer_size, pca_train_score, color='green', marker='o')
plt.plot(hidden_layer_size, pca_test_score, color='green', linestyle='dashed', marker='o')

plt.plot(hidden_layer_size, ica_train_score, color='blue', marker='o')
plt.plot(hidden_layer_size, ica_test_score, color='blue', linestyle='dashed', marker='o')

plt.plot(hidden_layer_size, rp_train_score, color='orange', marker='o')
plt.plot(hidden_layer_size, rp_test_score, color='orange', linestyle='dashed', marker='o')

plt.plot(hidden_layer_size, rf_train_score, color='purple', marker='o')
plt.plot(hidden_layer_size, rf_test_score, color='purple', linestyle='dashed', marker='o')

# Custom Legend
green = mlines.Line2D([], [], color='green', label='PCA')
blue = mlines.Line2D([], [], color='blue', label='ICA')
orange = mlines.Line2D([], [], color='orange', label='RP')
purple = mlines.Line2D([], [], color='purple', label='RF')
black = mlines.Line2D([], [], color='black', label='Train Score')
blackd = mlines.Line2D([], [], color='black', linestyle='dashed', label='Test Score')

plt.legend(handles=[green,blue,orange, purple, black, blackd], loc='best')
plt.grid()
plt.show()
```
